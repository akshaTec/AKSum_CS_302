```python
import matplotlib.pyplot as plt
import numpy as np

# Combat effectiveness parameters
f = b = 0.05

# Initial number of ships
F0 = 33
B0 = 27

# Time step
dt = 1

# Time range
t = np.arange(0, 10, dt)

# Initial conditions
B = np.zeros_like(t)
F = np.zeros_like(t)

B[0] = 13
F[0] = 3

# Reserve ships
B_reserve = 14

# Ships engaged in the first stage
B_engaged = 13
F_engaged = 3
breakflg = False

for i in range(1, len(t)):
  # Update French ships
  dF = -b * F[i-1] * dt
  F[i] = F[i-1] + dF
  if F[i] < 2:
    breakflg = True

  # Update British ships
  dB = -f * B[i-1] * dt
  B[i] = B[i-1] + dB
  
  print("t = ", t[i])
  print("B = ", B[i])
  print("F = ", F[i])

  # Stop integration when the number of French ships is between 1 and 2
  if breakflg:
    print("broken at t = ", t[i])
    break

plt.plot(t, B, label='British Ships')
plt.plot(t, F, label='French Ships')
plt.xlabel('Time')
plt.ylabel('Number of Ships')
plt.title('Battle of Trafalgar Simulation (First Stage)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=b= 0.05

F0=33
B0=27
```

```python
dt = 0.1


TLOW = 0
THIGH = 50

DT = int((1 / dt) *(THIGH - TLOW))

t = np.linspace(TLOW, THIGH, DT)
```

```python
f01=3
b01=13
```

```python
def f_dot(F,B):
    return -B*b

def b_dot(F,B):
    return -F*f
```

```python
def euler(t,dt, order):
    ans_f = [f01]
    ans_b = [b01]

    for i in range(1, DT):
        if order == 1:
            # first order
            ans_f.append(ans_f[i-1] + dt*f_dot(ans_f[i-1],ans_b[i-1]))
            ans_b.append(ans_b[i - 1] + dt * b_dot(ans_f[i-1],ans_b[i-1]))
            
        
    return ans_b, ans_f
```

```python
ans_b, ans_f = euler(t,dt,1)
```

```python
for i in range(len(ans_f)):
    if ans_f[i] < 1:
        ans_f = ans_f[:i]
        ans_b = ans_b[:i]
        t= t[:i]
        break
```

```python
ans_b1= ans_b
ans_f1= ans_f
t1= t
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('Time (days)')
plt.title("Battle of Trafalgar")
plt.ylabel('Ships Left')
plt.plot(t,ans_f, label='French')
plt.plot(t,ans_b, label='British')
plt.legend()
plt.savefig("q2_1.png")
plt.show()
```

```python
print(f"British: {ans_b[-1]} ships left") 
```

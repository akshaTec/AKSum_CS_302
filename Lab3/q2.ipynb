```python
def xdot_calc (X, A):
    # calculate list Xdot using list X and constant A
    Xdot = []
    for x in X:
        Xdot.append((x+A)*(1-x))
    return Xdot
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

A_values = [0, 0.2, 0.3]
x_values = np.linspace(0, 1, 100)

plt.figure(figsize=(10, 8))
for a in A_values:
    x_values = np.linspace(-a, 1, int((1+a)*100))
    Xdot = xdot_calc(x_values, a)
    plt.plot(x_values, Xdot, label='A = {}'.format(a))

# Plot the solution
plt.xlabel('X')
plt.ylabel('dX/dT')
plt.title('Dynamical Equation Solution')
plt.legend()
# plt.grid()
plt.savefig('dynamical_equation_solution.png')
plt.show()
```

```python
def calculate_value_np(A, T):
    numerator = 1 - np.exp(-(1 + A) * T)
    denominator = 1 + (1 / A) * np.exp(-(1 + A) * T)
    return numerator / denominator
```

```python
t_values = np.linspace(0, 10, 1000)
A_values = [0.01, 0.2, 0.3]

plt.figure(figsize=(10, 8))
for a in A_values:
    x_values = [0.0001] + [calculate_value_np(a, t) for t in t_values.tolist()[1:]]
    plt.plot(t_values, x_values, label='A = {}'.format(a))

# Plot the solution
plt.xlabel('T')
plt.ylabel('X(T)')
plt.title('Dynamical Equation Solution')
plt.legend()
# plt.grid()
plt.savefig('X(T) vs T.png')
plt.show()
```

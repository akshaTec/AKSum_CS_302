```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt= 0.01
n= 0.035
alpha= 0.5
s= 0.15
```

```python
def X_dot(x,y):
    return s*(x**alpha) - n*x

def Y_dot(x,y):
    return 1
```

```python
x0=0
y0=0

TLOW = 0
THIGH = 20

DT = int((1 / dt) *(THIGH - TLOW))

t = np.linspace(TLOW, THIGH, DT)
```

```python
x = np.linspace(TLOW, THIGH, DT)
xd= X_dot(x,0)
```

```python
def euler(t,dt, order):
    ans_x = [x0]
    ans_y = [y0]

    for i in range(1, DT):
        if order == 1:
            # first order
            ans_y.append(ans_y[i-1] + dt*Y_dot(ans_x[i-1],ans_y[i-1]))
            ans_x.append(ans_x[i - 1] + dt * X_dot(ans_x[i-1],ans_y[i-1]))
            
        
    return ans_x, ans_y
```

```python
#plot x_dot vs x
#also plot point where r_dot = 0
plt.figure(figsize=(10,8))
plt.plot(x,xd, label='r_dot vs r')
plt.title('r_dot vs r')
plt.xlabel('r')
plt.ylabel('r_dot')
plt.legend()
plt.savefig('q1_1.png')
plt.show()
```

```python
def y1(x):
    return s*(x**alpha)

def y2(x):
    return n*x
```

```python
y1 = y1(x)
y2 = y2(x)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, y1, label='y1 vs r')
plt.plot(x, y2, label='y2 vs r')
plt.title('y1 and y2 vs r')
plt.xlabel('r')
plt.ylabel('y')
plt.legend()
plt.savefig('q1_2.png')
plt.show()
```

```python
#integrate xdot using euler
r, y1 = euler(t, dt, 1)
```

```python
#plot xd and t 
plt.figure(figsize=(10,8))
plt.plot(t, r, label='r vs t')
plt.title('r vs t')
plt.xlabel('t')
plt.ylabel('r')
plt.legend()
# plt.savefig('q1_3.png')
plt.show()
```

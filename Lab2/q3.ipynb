```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import odeint
import math

from sympy import symbols, diff
import sympy
```

```python
dt = 0.01

x0 = 0
y0= 0

TLOW = 0
THIGH = 300

DT = int((1 / dt) *(THIGH - TLOW))

t = np.linspace(TLOW, THIGH, DT)
```

```python
k1= 0.6931
k2= 0.0231 
```

```python
def X_dot(x,y):
    return 1 -k1*x

def Y_dot(x,y):
    return k1*x - k2*y
```

```python
def euler(t, dt, order):
    ans_x = [x0]
    ans_y = [y0]

    for i in range(1, DT):
        if order == 1:
            # first order
            ans_y.append(ans_y[i-1] + dt*Y_dot(ans_x[i-1],ans_y[i-1]))
            ans_x.append(ans_x[i - 1] + dt * X_dot(ans_x[i-1],ans_y[i-1]))
            
        
    return ans_x, ans_y
```

```python
x,y = euler(t, dt, 1)
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('t (hours)')
plt.ylabel('x')
plt.title("X vs T")
plt.plot(t, x, label='x')
plt.legend()
plt.savefig("q3_1.png")
plt.show()
           
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('t (hours)')
plt.ylabel('y')
plt.title("Y vs T")
plt.plot(t, y, label='y')
plt.legend()
plt.savefig("q3_2.png")
plt.show()
```

```python
# take the max value of y and t at that point
max_y = max(y)
max_t = t[y.index(max_y)]
print(f'Max value of Y = {max_y}, at time= {max_t} hour')
```

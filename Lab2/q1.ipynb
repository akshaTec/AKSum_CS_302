```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cin= 3
v= 10**12
f= 5*10**8
```

```python
a= f*cin/v
b= f/v
c_o= 10
```

```python
def c_dot(a,b,c):
    return a - b*c
```

```python
k= -(1/b*np.log(b*c_o -a))
```

```python
def fc(a,b,k,t):
    return (a+ np.exp(-b*(t+k)))/(b)
```

```python
x= np.linspace(0,10**4,1000)
y= fc(a,b,k,x)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.xlabel("Time (s)") 
plt.ylabel("Concentration (M)")
plt.show()
```

```python
def time_for_some_amount(c_o,b,a,k):
    return (-1/b)*np.log(c_o*b-a) - k
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parameters
k = 1.2
l = 1.5
h = 0
g = 0
alpha = 1.8
beta = 1.7

# time step
dt = 0.01
t_max = 10

# store the solution
t = np.arange(0, t_max + dt, dt)
x = np.zeros_like(t)
y = np.zeros_like(t)

# initial conditions
x[0] = 0.5
y[0] = 0.9

# Euler integration
for i in range(1, len(t)):
    x[i] = x[i - 1] + dt * (k * y[i - 1] + g - alpha * x[i - 1])
    y[i] = y[i - 1] + dt * (l * x[i - 1] + h - beta * y[i - 1])

plt.figure(figsize=(10, 8))
plt.plot(t, x, label=r"$x(t)$")
plt.plot(t, y, label=r"$y(t)$")

plt.yscale('log')  # Change Y-axis to log scale

plt.xlabel("t")
plt.ylabel("x(t), y(t)")
plt.legend()
plt.grid()
plt.savefig("A1.png")

plt.show()
```

```python
plt.figure(figsize=(10, 8))

# Plot y vs x
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.savefig("A2.png")

plt.show()
```

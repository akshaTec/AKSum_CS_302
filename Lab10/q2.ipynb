```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A= 10**-6
B= 0.333
N= 10**6
a=b= 0.02
```

```python
def N_dot(N):
    return (b-a)*N
```

```python
def x_dot(x,y,z):
    return A*x*y - B*x -a*x

def y_dot(x,y,z):
    return -A*x*y +b*N -a*y

def z_dot(x,y,z):
    return B*x - a*z
```

```python
x0 = 10**5
y0 = 9 * 10**5
z0 = 0

# Define the time step
dt = 0.005

t0=0
t1= 170

t= np.arange(t0,t1+dt,dt)

x = np.zeros(len(t))
y = np.zeros(len(t))
z = np.zeros(len(t))

x[0] = x0
y[0] = y0
z[0] = z0

for i in range(1,len(t)):
    x[i] = x[i-1] + dt*x_dot(x[i-1],y[i-1],z[i-1])
    y[i] = y[i-1] + dt*y_dot(x[i-1],y[i-1],z[i-1])
    z[i] = z[i-1] + dt*z_dot(x[i-1],y[i-1],z[i-1])
```

```python
#plot x and t
plt.figure(figsize=(10,8))
plt.plot(t,x,label='x')
# plt.plot(t,z,label='z')
plt.legend()
plt.xlabel('Days')
plt.ylabel('No. of Individuals')
plt.title('Endemic diseases')
plt.savefig("q2_1.png")
plt.show()
```

```python
#get all the local peaks of x and get their time values

local_max = []
for i in range(1,len(x)-1):
    if x[i] > x[i-1] and x[i] > x[i+1]:
        local_max.append(t[i])

print(local_max)
```

```python
#plot y and z vs t
plt.figure(figsize=(10,8))
plt.plot(t,y,label='y')
plt.legend()
plt.xlabel('Days')
plt.ylabel('No. of Individuals')
plt.title('Endemic diseases')
plt.savefig("q2_2.png")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,z,label='z')
plt.legend()
plt.xlabel('Days')
plt.ylabel('No. of Individuals')
plt.title('Endemic diseases')
plt.savefig("q2_3.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A= 2.18 * 10**-3
B= 0.44
N= 763
```

```python
def x_dot(x, y,z):
    return A*x*y - B*x

def y_dot(x, y,z):
    return - A*x*y\
    
def z_dot(x, y,z):
    return B*x
```

```python
x0 = 1
y0 = N- x0
z0 = 0

# Define the time step
dt = 0.001

t0=0
t1=25

t= np.arange(t0,t1+dt,dt)

x = np.zeros(len(t))
y = np.zeros(len(t))
z = np.zeros(len(t))

x[0] = x0
y[0] = y0
z[0] = z0

for i in range(1,len(t)):
    x[i] = x[i-1] + dt*x_dot(x[i-1],y[i-1],z[i-1])
    y[i] = y[i-1] + dt*y_dot(x[i-1],y[i-1],z[i-1])
    z[i] = z[i-1] + dt*z_dot(x[i-1],y[i-1],z[i-1])
```

```python
#plot x and y wrt t
plt.figure(figsize=(10,8))
plt.plot(t,x,label='x')
plt.plot(t,y,label='y')
# plt.plot(t,z,label='z')
plt.legend()
plt.xlabel('Days')
plt.ylabel('No. of Individuals')
plt.title('Epidemics')
plt.savefig("q1_1.png")
plt.show()
```

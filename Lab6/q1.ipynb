```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read traffic_time.csv into a DataFrame
df = pd.read_csv('traffic_time.csv')
df
```

### The first column gives the average volume of traffic flowing towards the east. 
### The second column gives the hour of the day, with the 0 hour being set at midday, 12 pm. So all forenoon hours are negative.
### The third column gives the average volume of the traffic flowing towards the west. 

```python
plt.figure(figsize=(10,8))
plt.xlabel('time')
plt.ylabel('traffic')
plt.plot(df.iloc[:, 1],df.iloc[:, 0], label='to the east')
plt.plot(df.iloc[:, 1],df.iloc[:, 2], label='to the west')

plt.legend()
# plt.savefig("q1_1.png")
plt.show()
```

# with A=44.0, = 8.53, = 0.19and  = 0.09.

```python
A = 44.0
mu = 8.53
lam = 0.19
beta = -0.09

def N(t):
  return A * np.exp(-(lam*t - beta)**2) * (mu + t**2)

t_vals = np.linspace(-12, 12, 1000)
N_vals = N(t_vals)

plt.figure(figsize=(10,8))
plt.xlabel('time')
plt.ylabel('traffic')
# plt.plot(df.iloc[:, 1],df.iloc[:, 0], label='to the east')
plt.plot(df.iloc[:, 1],df.iloc[:, 2], label='to the west')



plt.plot(t_vals, N_vals)
plt.xlabel('t')
plt.ylabel('N(t)')
plt.title('Plot of N(t) vs t')
# plt.xlim(-12, 12)
# plt.ylim(0, 50)
plt.legend()
plt.savefig("q1_1.png")
plt.show()
```

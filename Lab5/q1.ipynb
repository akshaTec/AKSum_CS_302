```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
```

```python

# Load data from the file
data = np.genfromtxt('parato_in.csv' , delimiter=',')
```

```python
data
```

```python
# data[:,2]
```

```python
# Fit the power law function
def power_law(x, A, B, alpha):
    return A + B * x**(-alpha)
```

```python
x= data[:,1]
y= data[:,2]
```

```python
initial_guess = [60, 915, 1.25]
```

```python

# Fit the model
params, params_covariance = scipy.optimize.curve_fit(power_law, x, y , p0=initial_guess)
B = 329400
# Generate the curve
# x_curve = [1,10,100,1000]  # generating x values for curve
x_curve = np.logspace(0, 4, 25)  # generating x values for curve

y_curve = [60 + ((B) * x**(-1.25)) for x in x_curve]  # generating y values for curve
# 60 + B * x**(-1.25)
y_curve
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_curve, y_curve, color='black', label='Power Law Fit')
plt.xscale('log')
plt.yscale('log')
plt.title('Pareto Distribution of Wealth in India')
plt.xlabel('Wealth (in M$)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy 
import pandas as pd
```

```python
gdb= pd.read_csv("india_gdp.csv", header=None, sep=',')
gdb[:10]
```

```python
#first col: year
#second col: gdp
#thrird col: trade
```

```python
plt.figure(figsize=(10,6))
plt.plot(gdb[0]+1960,gdb[1], label='GDP')
plt.title('GDP of India')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.yscale('log')
plt.show()
```

```python
def logictic_f(k,t,a,x0):
    return k*x0*np.exp(a*t)/(k+x0*(np.exp(a*t)-1))
```

```python
india_gdp_model = logictic_f(6*10**12, gdb[0], 0.08,3.6*10**10)
```

```python
plt.figure(figsize=(10,6))
plt.plot(gdb[0]+1960,gdb[1], label='GDP')
plt.plot(gdb[0]+1960,india_gdp_model, label='GDP model')
plt.title('GDP of India')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
def fun(x, a, k, x0):
    ret = 1960 + (1/a) * np.log((x*(k -x0)) / (x0*(k-x)))
    return ret
```

```python
pred= fun(5*10**12,0.08,6*10**12,3.6*10**10)
print(f"India's gdp will touch 4 trillion by {pred} ")
```

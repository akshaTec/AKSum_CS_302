```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
import pandas as pd

df= pd.read_csv("india_gdp.csv", header=None, sep=',')
# df.columns = ['Year', 'GDP', 'Trade']

print(df.head())
```

```python
def logictic_f(k,t,a,x0):
    return k*x0*np.exp(a*t)/(k+x0*(np.exp(a*t)-1))
```

```python
india_gdp_model = logictic_f(6*10**12, df[0], 0.08,3.6*10**10)

plt.figure(figsize=(10,6))
plt.plot(df[0]+1960,df[1], label='GDP', color='red', marker='o', linestyle='none', markersize=5)
plt.plot(df[0]+1960,india_gdp_model, label='GDP model',color='darkred')
name='india'
################
A = 0.10   
K = 3     # in trillions
X0 = 0.4   # in 10 billions(e+10) predicted


trade_model = logictic_f(K*10**12, df[0], A,X0*10**10)

plt.plot(df[0]+1960,df[2], label='Trade', color='lime', marker='o', linestyle='none', markersize=5)
plt.plot(df[0]+1960,trade_model, label='Trade model', color='darkgreen')
################
plt.title(f'{name}'.upper())
plt.xlabel('Year')
plt.ylabel('GDP & Trade')

plt.yscale('log')
plt.legend()
# plt.grid()
plt.savefig(f'{name}.png', dpi=500)
plt.show()
```

```python
def fun(x, a, k, x0):
    ret = 1960 + (1/a) * np.log((x*(k -x0)) / (x0*(k-x)))
    return ret
```

```python
pred= fun(4*10**12,0.08,6*10**12,3.6*10**10)
print(pred)
```

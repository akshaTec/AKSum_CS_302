```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
import pandas as pd

df= pd.read_csv("india_gdp.csv", header=None, sep=',')
# df.columns = ['Year', 'GDP', 'Trade']

print(df.head())
```

```python
def logictic_f(k,t,a,x0):
    return k*x0*np.exp(a*t)/(k+x0*(np.exp(a*t)-1))
```

```python
india_gdp_model = logictic_f(6*10**12, df[0], 0.08,3.6*10**10)

# plt.figure(figsize=(10,6))
plt.plot(df[0]+1960,df[1], label='GDP')
plt.plot(df[0]+1960,india_gdp_model, label='GDP model')

plt.title('GDP of India')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```

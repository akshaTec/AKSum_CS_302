```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=0.21827
C= 0.06069
B=0
D=0
alpha= 0.05289
beta= 0.00459
```

```python
def x_dot(x,y):
    return A*x- B*x*x - alpha*x*y

def y_dot(x,y):
    return C*y - D*y*y - beta*x*y
```

```python
# Define the initial values
x0 = 0.5
y0 = 1.5

# Define the time step
dt = 0.0001

t0=0
t1=50

t= np.arange(t0,t1+dt,dt)

x = np.zeros(len(t))
y = np.zeros(len(t))

x[0] = x0
y[0] = y0

for i in range(1,len(t)):
    x[i] = x[i-1] + dt*x_dot(x[i-1],y[i-1])
    y[i] = y[i-1] + dt*y_dot(x[i-1],y[i-1])
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,x,label='SpeciesX')
plt.plot(t,y,label='SpeciesY')
plt.title('Competitive exclusion Model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.savefig("q1_1.png")
plt.show()

```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.title('Competitive exclusion Model')
plt.xlabel('Population of SpeciesX')
plt.ylabel('Population of SpeciesY')
plt.legend()
plt.savefig("q1_2.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A= 1.0
B= 0.01
C= 0.5
D= 0.005
e= 0
```

```python
def x_dot(x,y):
    return A*x - B*x*y - e*x

def y_dot(x,y):
    return -C*y + D*x*y - e*y
```

```python
x0 = 200
y0 = 80

# Define the time step
dt = 0.001

t0=0
t1=20

t= np.arange(t0,t1+dt,dt)

x = np.zeros(len(t))
y = np.zeros(len(t))

x[0] = x0
y[0] = y0

for i in range(1,len(t)):
    x[i] = x[i-1] + dt*x_dot(x[i-1],y[i-1])
    y[i] = y[i-1] + dt*y_dot(x[i-1],y[i-1])
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,x,label='Pray')
plt.plot(t,y,label='Predator')
plt.title('Predator-prey dynamics')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.savefig("q2_1.png")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.title('Predator-prey dynamics')
plt.xlabel('Population of Pray')
plt.ylabel('Population of Predator')
plt.legend()
plt.savefig("q2_2.png")
plt.show()
```

```python
print("Max population of Pray",max(x))
print("Max population of Predator",max(y))
```
